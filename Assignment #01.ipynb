#### 5.Perform meaningful visualizations using the dataset.

### 1.Bar plot

```python
plt.bar(data['Classification'],data['SL'])
plt.title('Iris Classification with SL')
plt.ylabel('SL')
plt.show()
```

```python
plt.bar(data['Classification'],data['PL'])
plt.title('Iris classification with PL')
plt.ylabel('PL')
plt.show()
```

### 2.Pairplot

```python
sns.pairplot(data)
```

### 3. Boxplot

```python
plt.boxplot(data['SL'])
plt.show()
```

### 4.Countplot

```python
sns.countplot(x=data.PW)
plt.show()
```

### 5.Scatterplot

```python
plt.scatter(data['SL'],data['SW'],s=10,color='blue')
```

### 6. Histogram

```python
plt.figure(figsize=(10,5))
plt.hist(data['PW'],color='green',rwidth=0.98,cumulative=True)
plt.title('histogram of PW',fontsize=20)
plt.xlabel('PW')
plt.show
```
